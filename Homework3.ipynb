```python
import pandas as pd
from pandas.tools.plotting import scatter_matrix
%matplotlib inline

#Read iris.txt into a data frame (2 point)
df = pd.read_csv('iris.txt')

#Data gets read into a dataframe

```

```python
#Print summary statistics for the data frame (both numeric and categorical columns) (2 point)
df.describe()

#Basic statistical info combining the species togethere
```

```python
#Generate a box plot for the data set (2 point)
df.boxplot(by='Species')

#Most interesting plot.  Shows clear difference between the species in petal length, sepal length, and petal width.  
#Sepal width falls in a narrow range across the species.
```

```python
#Generate a line plot the dataset (2 point)
#df.plot()
df.plot(subplots = 'True')

#Lines plots the data but ins't as useful all the boxplot
```

```python
#Generate a histogram of the dataset (2 point)
df.plot.hist()

#Shows the historigram of values for the different measures.  Still like the boxplot best
```

```python
#Print the summary statistics for each species in the dataset (2 point)
df.groupby(['Species']).describe()

#The group by describe is much more useful than our initial summary statistics because it is 
#broken out by species. 
```

```python
#Generate a boxplot for each species in the dataset (2 point)
df.boxplot(by='Species')

#My favorite, boxplot by species.  Gives the best idea of difference of measures by species.
```

```python
#Generate a scatter matrix for the dataset (2 point)
#df.plot.scatter(x='SepalLength', y='SepalWidth');
df.plot.scatter(x='SepalLength', y='PetalLength');

#Plot sepalength vs petal length looking for difference
```
